```python
import pandas as pd
import re
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
import nltk
```

```python
# Download stopwords
nltk.download('stopwords')
```

```python
# Function to clean and preprocess text data
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove non-alphabetic characters using regex
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text
```

```python
# Load the dataset
df = pd.read_csv('/content/Nlpdataset2.csv')
df.head()
```

```python
# Assume the text data is in a column called 'text_column', change if needed
df['cleaned_text'] = df['Review'].apply(preprocess_text)
df.head()
```

```python
# Create an instance of CountVectorizer with binary=True to ensure binary output (1 for presence, 0 for absence)
vectorizer = CountVectorizer(binary=True)

# Fit the vectorizer to the cleaned text data and transform it
bow_matrix = vectorizer.fit_transform(df['cleaned_text'])

# Convert the Bag-of-Words matrix into a DataFrame
bow_df = pd.DataFrame(bow_matrix.toarray(), columns=vectorizer.get_feature_names_out(), index=df.index)

# Display the binary Bag-of-Words DataFrame
print(bow_df)
```
