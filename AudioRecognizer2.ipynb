```python
import speech_recognition as sr
from pydub import AudioSegment
from google.colab import files
import os
```

```python
# Initialize the recognizer
recognizer = sr.Recognizer()
```

```python
def upload_audio_file():
    # Use Google Colab's file upload tool
    uploaded = files.upload()
    for file_name in uploaded.keys():
        print(f"Uploaded file: {file_name}")
        return file_name
```

```python
def capture_audio_from_file(file_path):
    # Check if the file format is not in WAV format and convert it to WAV
    if not file_path.endswith(".wav"):
        audio = AudioSegment.from_file(file_path)
        # Convert the file to WAV format and save it temporarily
        file_path = "temp.wav"
        audio.export(file_path, format="wav")

    # Use speech_recognition to process the WAV file
    with sr.AudioFile(file_path) as source:
        print("Reading audio file...")
        audio = recognizer.record(source)
    return audio
```

```python
def convert_voice_to_text(audio):
    try:
        text = recognizer.recognize_google(audio)
        print("You said: " + text)
    except sr.UnknownValueError:
        text = ""
        print("Sorry, I didn't understand that.")
    except sr.RequestError as e:
        text = ""
        print("Error; {0}".format(e))
    return text
```

```python
def process_voice_command(text):
    if "hello" in text.lower():
        print("Hello! How can I help you?")
    elif "goodbye" in text.lower():
        print("Goodbye! Have a great day!")
        return True
    else:
        print("I didn't understand that command. Please try again.")
    return False
```

```python
def main():
    # Upload the audio file using Colab's file upload
    file_path = upload_audio_file()

    # Process the audio file
    audio = capture_audio_from_file(file_path)
    text = convert_voice_to_text(audio)
    process_voice_command(text)

# Run the main function
if __name__ == "__main__":
    main()
```
