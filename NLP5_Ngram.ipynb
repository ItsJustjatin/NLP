```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
import nltk
from collections import Counter
```

```python
# Ensure you have the necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Load your dataset
df = pd.read_csv('/content/Nlpdataset2.csv')
df.head()
```

```python
# Example: Assume the text is in a column named 'text'
text_data = df['Review'].dropna().tolist()  # Convert the column to a list
```

```python
# Tokenize the sentence into words
words = []
for text in text_data:
    words.extend(text.split())

# Create bigrams from the list of words
unigrams = ngrams(words, 1)

# Print the bigrams
for unigram in unigrams:
    print(unigram)
```

```python
# Tokenize the sentence into words
words = []
for text in text_data:
    words.extend(text.split())

# Create bigrams from the list of words
bigrams = ngrams(words, 2)

# Print the bigrams
for bigram in bigrams:
    print(bigram)
```
